```python
from netCDF4 import Dataset
import os

user = os.environ['USERPROFILE']

# Define o caminho do arquivo
file_path = os.path.join(user, 'Documentos', 'emit', 'EMIT_L2A_RFL_001_20230119T114235_2301907_004.nc')

# Abre o arquivo nc para leitura
nc_file = Dataset(file_path, 'r')  # 'r' significa que o arquivo será aberto no modo de leitura

# Mostra informações sobre o arquivo
print("Informações sobre o arquivo:", nc_file)

# Lista todos os nomes de variáveis
print("Nomes das variáveis:", nc_file.variables.keys())

# Lista todas as dimensões do arquivo
print("Dimensões do arquivo:", nc_file.dimensions.keys())

# Lista os atributos globais do arquivo
print("Atributos globais do arquivo:", nc_file.ncattrs())


# Fecha o arquivo
# nc_file.close()
```

```python
# Plotagem de dados
import matplotlib.pyplot as plt
import numpy as np

# Define a variável a ser plotada
var = nc_file.variables['reflectance']

# Define a dimensão da variável
dim = var.dimensions

# Define a escala de cores
cmap = 'jet'

# Define o título do gráfico
title = 'RFL'

# Define o nome do eixo x
x_label = '200px'
# Define o nome do eixo y
y_label = '200px'

# Define o tamanho da figura
fig = plt.figure(figsize=(10, 10))

# Define o tamanho da fonte
plt.rcParams.update({'font.size': 12})

# Define o título da figura
plt.title(title)

# Define o nome do eixo x
plt.xlabel(x_label)
# Define o nome do eixo y
plt.ylabel(y_label)

# Define o intervalo de cores
plt.imshow(var[:,:,0], cmap=cmap)  

# Define a barra de cores
plt.colorbar()

# Legenda
plt.legend()

# Mostra a figura
plt.show()

# Fecha o arquivo
nc_file.close()
```
